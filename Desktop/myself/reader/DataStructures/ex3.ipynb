```python
#3-2 code section

def prefix_average(s):
    """Return list such that, for all j, A[j] equals average of s[0], s[1], s[2], ....."""
    n = len(s)
    a = [0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += s[i]
        a[j] = total / (j + 1)
    return a
```

```python
# code 3-3 O(n ** 2)

def prefix_average2(s):
    n = len(s)
    a = [0] * n
    for j in range(n):
        a[j] = sum(s[0:j+1]) / (j+1)
    return a
```

```python
def unique1(s):
    for j in range(len(s)):
        for k in range(j+1, len(s)):
            if s[j] == s[k]:
                return False
        return True
```

#### #3-8

```python
def unique2(s):
    temp = sorted(s)
    for j in range(1, len(temp)):
        if temp[j-1] == temp[j]:
            return False
    return True
```
